```python
import requests
from fake_useragent import UserAgent
from lxml import etree
import random
import time
```

# 进行爬取数据前的反爬处理

```python
ua = UserAgent()
with open(r'C:\Users\SilentCat\ip.txt','r')as f: #该文件为在快代理网站下爬取的代理地址
    data = f.readlines()
    f.close()
reg = []
for i in data:
    k = eval(i)
    reg.append(k)
ip = random.choice(reg)
proxies = {
    "http":'ip'
}
headers = {
    'User-Agent':ua.random, #随机生成ua进行ua伪装
    'Referer':'https://movie.douban.com/top250?start=0&filter='
}
```

# 定义一个函数爬取豆瓣top250页面中相应电影的详细网址

```python
def get_detail_url(url):
    response = requests.get(url=url,headers=headers,proxies=proxies)
    if response.status_code == 200:
        page_text = response.text
        html = etree.HTML(page_text)
        detail_url = html.xpath('.//div[@class="hd"]/a/@href')
        print(detail_url)
        return detail_url
    else:
        return False 
```

# 定义一个爬取相应电影数据d

```python
def parse_detail_url(detail_url):
    response = requests.get(url=detail_url,headers=headers,proxies=proxies)
    if response.status_code == 200:
        page_text = response.text
        html = etree.HTML(page_text)
        datalist = []
        #电影标题
        Movie_title = html.xpath('.//h1/span[@property="v:itemreviewed"]/text()')[0]
        print(Movie_title)
        #导演
        Director = html.xpath('//div[@id="info"]/span[1]/span[2]/a/text()')
        #上映年份
        Year = html.xpath('.//*[@id="content"]/h1/span[2]/text()')
        print(Year)
        #豆瓣评星
        Score = html.xpath('.//div[@id="interest_sectl"]//strong[@class="ll rating_num"]/text()')
        #影评人数
        num_review = html.xpath('.//div[@id="interest_sectl"]//div[@class="rating_sum"]/a/span/text()')[0] + '人'
        #剧情简介
        Introduction = html.xpath('.//*[@id="link-report"]/span[2]/text()[1]')
        if len(Introduction) == 0:
            Introduction = html.xpath('.//*[@id="link-report"]/span[1]/text()[1]')
        with open("top250_test6.json",'a+',encoding='utf-8')as f:
                f.write("影片名称：%s,导演：%s,上映年份：%s,评分：%s,影评人数：%s,剧情简介：%s"%( Movie_title, Director,Year,Score,num_review,Introduction))
                f.write("\n")
                f.write("============================\n")
                f.close()
        time.sleep(2)
        
    else:
        return False
```

```python
def snatch():
    url = 'https://movie.douban.com/top250?start={}&filter='
    for i in range(10):
        top250_url = url.format(i*25)
        detail_urls = get_detail_url(top250_url)
        if detail_urls == 0:
            print("出错啦\n")
            return False
        for detail_url in detail_urls:
            parse_detail_url(detail_url)
        print("一页数据已经爬取完了\n")
        time.sleep(40)
        
```

```python
if __name__ == "__main__":
    snatch()
   
```
